# Organize Data

```python
import scanpy as sc
import pandas as pd
from sklearn.cluster import KMeans
import anndata as ad
```

```python
# Cluster source cells
fname = '/nfs/turbo/umms-indikar/shared/projects/geneformer/fib15k/originals/2024-07-31_07-01-47_job_number_83.h5ad'
adata = sc.read_h5ad(fname)

# Get the data matrix
data_matrix = adata.X

# Perform KMeans clustering
kmeans = KMeans(n_clusters=25, random_state=0)
kmeans.fit(data_matrix)

# Get the centroids of each cluster
cluster_centroids = kmeans.cluster_centers_
```

```python
# Create an AnnData object for the centroids of the current file
adata_centroids = ad.AnnData(X=cluster_centroids)
adata_centroids.obs['recipe'] = 'source'
adata_centroids.obs['filename'] = fname
adata_fib = adata_centroids
adata_fib.obs
```

```python
# Cluster target cells
fname = '/nfs/turbo/umms-indikar/shared/projects/geneformer/fib15k/working_output/hsc_1.h5ad'
adata = sc.read_h5ad(fname)

# Get the data matrix
data_matrix = adata.X

# Perform KMeans clustering
kmeans = KMeans(n_clusters=25, random_state=0)
kmeans.fit(data_matrix)

# Get the centroids of each cluster
cluster_centroids = kmeans.cluster_centers_
```

```python
# Create an AnnData object for the centroids of the current file
adata_centroids = ad.AnnData(X=cluster_centroids)
adata_centroids.obs['recipe'] = 'target'
adata_centroids.obs['filename'] = fname
adata_hsc = adata_centroids
adata_hsc.obs
```

```python
# Combine HSC targets with file of reprogrammed embeddings, add meta info, and save
fname = '/nfs/turbo/umms-indikar/shared/projects/geneformer/fib15k/jpic/cluster_centroids.h5ad'
adata_reprogrammed = sc.read_h5ad(fname)
adata = ad.concat([adata_hsc, adata_fib, adata_reprogrammed])
unique_genes = []
for i in adata.obs['recipe'].values:
    unique_genes += i.split(';')
unique_genes = list(set(unique_genes))
unique_genes

for gene in unique_genes:
    vals = []
    for i in adata.obs['recipe'].values:
        vals.append(gene in i)
    adata.obs[gene] = vals
adata.write('/nfs/turbo/umms-indikar/shared/projects/geneformer/fib15k/jpic/HSC-reprogrammed-DE.h5ad')
adata.obs
```

```python
color_recipes = ['target', 'source']

# Scale the data and perform PCA
sc.pp.scale(adata, max_value=10)
sc.tl.pca(adata, svd_solver='arpack')

# Compute the neighborhood graph
sc.pp.neighbors(adata, n_neighbors=15, n_pcs=40)

# Compute the UMAP representation
sc.tl.umap(adata)

# Add a column to indicate whether the recipe matches the target
adata.obs['is_target_or_source'] = adata.obs['recipe'].isin(color_recipes)

# Plot the UMAP, coloring by the new column
sc.pl.umap(adata, color='is_target_or_source')
```
