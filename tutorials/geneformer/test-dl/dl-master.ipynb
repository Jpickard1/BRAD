# Digital laboratory

The purpose of this file is to demonstrate the digital library framework consisting of multiple BRADs.

```python
import os
import sys
import numpy as np
import pandas as pd
from BRAD import brad
from BRAD import llms
```

```python
llm = llms.load_nvidia(model_name='google/gemma-2-9b-it')
brad_llm = llms.load_nvidia()
agents = [
    brad.chatbot(interactive=False, llm=brad_llm),
    brad.chatbot(interactive=False, llm=brad_llm)
]
```

```python
llm.invoke('what day is today').content
```

```python
history = """System: You are an AI in charge of managing a team of AI research assitants. You will
prompt BOTS 1 and 2 with instructions in order and take the responses from them. Each response will
be recorded in the chat history, and you must manage the flow of information between the BOTS. If
the chat history is short, the conversation has not begun yet and you must initiate it.

Human Instructions: Have two bots communicate to discuss the role of Transcription Factors in
cellular reprogramming.

Chat History:
"""

manager_prompt = """Given the chat history, construct a set of instructions for BOT {botnum} to
be executed next. Do not include any informationnot relevant to BOT {botnum} in your response.
"""
# and please provide your output in the format of:
# Instructions=


cont=True
count = 0
while count < 2:
    count += 1
    for i in range(1, 1 + len(agents)):

        # Read history and determine what to ask next
        instructions = history + manager_prompt.format(botnum=i)

        print(f"{instructions=}")
        botinstructions = llm.invoke(instructions).content
        print(f"{botinstructions=}")

        # invoke next agent
        response = agents[i - 1].invoke("/force RAG " + botinstructions)
        print(f"{response=}")

        # botinstructions = "YYY"
        # response = "XXX"
        
        history += "Instructions: " + botinstructions + "\n\n"
        history += "Bot " + str(i) + ": " + response + "\n\n"
        
        print('\n\n')
    print("****************************************************")
    print("****************************************************")
    print(history)
    print("****************************************************")
```
