````python
from langchain.llms import OpenAI
from langchain.chains import ConversationChain
from langchain.prompts import PromptTemplate
from langchain.memory import ConversationBufferMemory

# Define a template for your conversation prompt
template = """
The following is a conversation between a human and an AI.
AI should be helpful, knowledgeable, and friendly.

History: {history}

Human: {input}
AI:
"""

# Initialize the prompt template
PROMPT = PromptTemplate(
    input_variables=["input", "history"],
    template=template,
)

# Initialize OpenAI LLM (assuming gpt-3.5-turbo)
llm = OpenAI(temperature=0.7,
             model='gpt-4o')

# Initialize conversation memory
memory = ConversationBufferMemory()

# Initialize the conversation chain with the prompt, LLM, and memory
conversation = ConversationChain(
    prompt=PROMPT,
    llm=llm,
    verbose=True,
    memory=memory
)

# Input prompt from user
input_text = "What is the capital of France?"

# Make a prediction
response = conversation.predict(input=input_text)

# Extract token usage from the last LLM call
token_usage = llm.get_last_token_usage()  # This function depends on LLM's API support

# Display the response
print(f"AI Response: {response}")

# Print token usage
print(f"Total tokens used: {token_usage['total_tokens']}")
print(f"Prompt tokens: {token_usage['prompt_tokens']}")
print(f"Completion tokens: {token_usage['completion_tokens']}")
````

````python
from langchain.llms import OpenAI
from langchain.chains import ConversationChain
from langchain.prompts import PromptTemplate
from langchain.memory import ConversationBufferMemory

# Define a template that includes both the conversation history and user input
template = """
The following is a conversation between a human and an AI.
AI should be helpful, knowledgeable, and friendly.

{history}

Human: {input}
AI:
"""

# Initialize the prompt template
PROMPT = PromptTemplate(
    input_variables=["history", "input"],  # Expecting both history and input
    template=template,
)

# Initialize OpenAI LLM (assuming gpt-3.5-turbo)
llm = OpenAI(model="gpt-3.5-turbo", temperature=0.7)

# Initialize conversation memory (which provides 'history' automatically)
memory = ConversationBufferMemory()

# Initialize the conversation chain with the prompt, LLM, and memory
conversation = ConversationChain(
    prompt=PROMPT,
    llm=llm,
    verbose=True,
    memory=memory
)

# Input prompt from user
input_text = "What is the capital of France?"

# Make a prediction
response = conversation.predict(input=input_text)

# Extract token usage from the last LLM call
token_usage = llm.get_last_token_usage()  # This function depends on LLM's API support

# Display the response
print(f"AI Response: {response}")

# Print token usage
print(f"Total tokens used: {token_usage['total_tokens']}")
print(f"Prompt tokens: {token_usage['prompt_tokens']}")
print(f"Completion tokens: {token_usage['completion_tokens']}")
````
